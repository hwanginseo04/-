```python
import seaborn as sns

# 데이터 로드
tips = sns.load_dataset('tips')

# 통계량 계산
tip_mean = tips['tip'].mean()     # 평균
tip_var = tips['tip'].var()       # 분산
tip_std = tips['tip'].std()       # 표준편차
tip_max = tips['tip'].max()       # 최댓값
tip_min = tips['tip'].min()       # 최솟값

print(f"평균(Mean): {tip_mean:.2f}")
print(f"분산(Variance): {tip_var:.2f}")
print(f"표준편차(Std Dev): {tip_std:.2f}")
print(f"범위: {tip_min} ~ {tip_max}")
```

평균 ($\mu$): 약 3.00 달러손님들이 평균적으로 3달러 정도의 팁을 낸다는 뜻입니다. 전체 결제 금액(total_bill)의 약 15~16% 수준입니다.

분산 ($\sigma^2$): 약 1.91데이터가 평균으로부터 얼마나 퍼져 있는지를 나타냅니다. 분산이 1.91이라는 것은 팁 금액들이 3달러 근처에 꽤 모여 있으면서도, 일부 큰 금액들로 인해 어느 정도 퍼짐성이 존재함을 의미합니다.

표준편차 ($\sigma$): 약 1.38 달러분산에 루트를 씌운 값으로, 실제 팁 금액이 평균에서 보통 1.38달러 정도 위아래로 왔다 갔다 한다고 이해하면 쉽습니다.

```python
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 5))

# 히스토그램과 밀도 그래프(KDE) 결합
sns.histplot(tips['tip'], kde=True, color='skyblue')

# 평균선 표시
plt.axvline(tip_mean, color='red', linestyle='--', label=f'Mean: {tip_mean:.2f}')
plt.title('Distribution of Tip Amounts')
plt.legend()
plt.show()
```

"2~3달러가 대세지만, 가끔 튀는 고액 팁이 섞여 있는 오른쪽으로 치우친 분포 입니다.
