```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# 1. 모든 경우의 수 생성 (주사위 2개)
dice1 = np.arange(1, 7)
dice2 = np.arange(1, 7)

sums = []
for d1 in dice1:
    for d2 in dice2:
        sums.append(d1 + d2)

# 2. PMF 계산
count = Counter(sums)
total = len(sums)

x = np.array(sorted(count.keys()))  # 가능한 값 (2~12)
pmf = np.array([count[i] / total for i in x])

# 3. CDF 계산
cdf = np.cumsum(pmf)

# 출력 확인
print("값 (X):", x)
print("PMF:", pmf)
print("CDF:", cdf)

# 4. 그래프
plt.figure()

# PMF
plt.subplot(1, 2, 1)
plt.bar(x, pmf)
plt.title("PMF (확률질량함수)")
plt.xlabel("합")
plt.ylabel("확률")
plt.xticks(np.arange(2, 13, 2)) # x축 눈금을 2단위로 설정

# CDF
plt.subplot(1, 2, 2)
plt.step(x, cdf, where='post')
plt.title("CDF (누적분포함수)")
plt.xlabel("합")
plt.ylabel("누적확률")
plt.xticks(np.arange(2, 13, 2)) # x축 눈금을 2단위로 설정

plt.tight_layout()
plt.show()
```

### 베르누이 분포 시뮬레이션 (성공 확률 0.5, 100회 실행)

성공 확률 `p=0.5`인 베르누이 시행을 100번 반복하여 결과를 시뮬레이션하고 시각화합니다.

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# 시뮬레이션 설정
p_success = 0.5  # 성공 확률
num_trials = 100 # 실행 횟수

# 베르누이 시행 시뮬레이션
# np.random.choice([0, 1], size=num_trials, p=[1-p_success, p_success])
# 또는 np.random.binomial(n=1, p=p_success, size=num_trials)
bernoulli_results = np.random.choice([0, 1], size=num_trials, p=[1 - p_success, p_success])

# 결과 집계
result_counts = Counter(bernoulli_results)

# 시각화를 위한 데이터 준비
labels = ['실패 (0)', '성공 (1)']
counts = [result_counts[0], result_counts[1]]

# 시각화
plt.figure(figsize=(7, 5))
plt.bar(labels, counts, color=['salmon', 'lightgreen'])
plt.title(f'베르누이 시행 결과 (성공 확률={p_success}, {num_trials}회)')
plt.xlabel('결과')
plt.ylabel('횟수')
plt.ylim(0, num_trials) # y축 범위를 0부터 총 시행 횟수까지로 설정
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 그래프에 총 실행 횟수 텍스트 추가
plt.text(0.5, num_trials * 0.9, f'총 실행 횟수: {num_trials}회', ha='center', va='center', fontsize=12, color='darkred')

plt.show()

print(f"총 {num_trials}회 시행 결과:")
print(f"실패 (0) 횟수: {result_counts[0]}회")
print(f"성공 (1) 횟수: {result_counts[1]}회")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# 설정
n = 10
p = 0.5
num_trials = 100

# 실험
results = []
for _ in range(num_trials):
    trial = bernoulli.rvs(p, size=n)
    results.append(np.sum(trial))

# 그래프
plt.figure()

# bins 핵심 수정 👇 (정수 하나씩 나오게)
bins = np.arange(-0.5, n + 1.5, 1)

plt.hist(results, bins=bins, density=True, alpha=0.7, edgecolor='black')

plt.xticks(range(0, n+1))  # x축을 0~10 모두 표시

plt.title("Binomial Distribution via Bernoulli (100 trials)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# 이항 분포 시뮬레이션 설정
p_success = 0.5  # 각 베르누이 시행의 성공 확률
n_bernoulli_trials_per_experiment = 100  # 각 이항 실험 내 베르누이 시행 횟수 (예: 동전을 100번 던짐)
num_binomial_experiments = 100 # 이항 실험을 반복할 횟수

# 이항 분포 시뮬레이션
# np.random.binomial(n, p, size)는 n번의 베르누이 시행에서 성공한 횟수를 size번 반복하여 반환합니다.
binomial_results = np.random.binomial(
    n=n_bernoulli_trials_per_experiment,
    p=p_success,
    size=num_binomial_experiments
)

# 결과 집계
success_counts = Counter(binomial_results)

# 시각화를 위한 데이터 준비
# 가능한 성공 횟수 (0부터 n_bernoulli_trials_per_experiment까지)
x_values = np.arange(0, n_bernoulli_trials_per_experiment + 1)
y_values = [success_counts.get(i, 0) for i in x_values]

# 시각화
plt.figure(figsize=(12, 6))
plt.bar(x_values, y_values, color='skyblue')
plt.title(f'이항 분포 시뮬레이션 (n={n_bernoulli_trials_per_experiment}, p={p_success}, {num_binomial_experiments}회 반복)')
plt.xlabel('성공 횟수')
plt.ylabel('빈도')
# x축 눈금이 너무 많아지는 것을 방지하기 위해 10단위로 설정
plt.xticks(np.arange(0, n_bernoulli_trials_per_experiment + 1, 10))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print(f"이항 실험 {num_binomial_experiments}회 반복 결과:")
# 모든 성공 횟수를 출력하면 너무 길어질 수 있으므로, 일부만 출력하거나 요약할 수 있습니다.
# 여기서는 모두 출력합니다.
for success_num in sorted(success_counts.keys()):
    print(f"성공 횟수 {success_num}: {success_counts[success_num]}회")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# 설정
n = 10
p = 0.5
num_trials = 1490

results = []

for _ in range(num_trials):
    trial = bernoulli.rvs(p, size=n)   # 0 or 1
    heads = np.sum(trial)
    tails = n - heads

    value = heads - tails   # 🔥 핵심 (갈톤보드 방식)
    results.append(value)

# bins 설정 (-5 ~ 5)
bins = np.arange(-n-1, n+2, 2) / 2  # 깔끔하게 정수 맞추기

plt.figure(figsize=(8,5))
plt.hist(results, bins=np.arange(-5.5, 6.5, 1), edgecolor='black')

plt.xticks(range(-5, 6))

plt.title("Galton Board Simulation (10 steps, 1490 trials)")
plt.xlabel("Heads - Tails")
plt.ylabel("Frequency")

plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. 실험 설정
n_coins = 10      # 동전 개수
n_trials = 149    # 실험 반복 횟수
offset = 5        # 차감할 값

# 2. 시뮬레이션 수행
# 0(뒷면)과 1(앞면)을 무작위로 생성 (n_trials행, n_coins열)
results = np.random.randint(0, 2, size=(n_trials, n_coins))

# 0을 -1(T)로 변환하고 1은 1(H)로 유지
results = np.where(results == 0, -1, 1)

# 각 시행별 합계 구하기 및 5 차감
final_values = results.sum(axis=1) - offset

# 3. 데이터 시각화 (이미지와 유사한 형태)
plt.figure(figsize=(10, 6))
sns.histplot(final_values, kde=True, bins=10, color='#5D99C6', edgecolor='black')

plt.title(f'Binomial Distribution (n={n_coins}, p=0.5, Trials={n_trials})')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.3)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 중심극한정리 시각화
plt.figure(figsize=(15, 5))

# 균등 분포에서 10,000개 추출 (Original)
data = np.random.uniform(0, 1, 10000)

# 다양한 표본 크기(n)에 따른 평균의 분포
for i, n in enumerate([1, 2, 30], 1):
    # n개씩 뽑아서 평균을 내는 과정을 1000번 반복
    sample_means = [np.mean(np.random.choice(data, n)) for _ in range(1000)]

    plt.subplot(1, 3, i)
    sns.histplot(sample_means, kde=True, color='skyblue')
    plt.title(f'Sample Size n={n}')
    if n == 30:
        plt.annotate('Normal Distribution!', xy=(0.5, 3), xytext=(0.7, 10),
                     arrowprops=dict(facecolor='black', shrink=0.05))

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 설정
n_size = 30       # 한 번에 뽑는 표본의 크기
n_iterations = 100 # 반복 횟수 (평균을 낼 횟수)

# 2. 중심극한정리 시뮬레이션
# 0~1 사이의 균등 분포(Uniform)에서 데이터를 추출합니다.
sample_means = []

for _ in range(n_iterations):
    # 0~1 사이에서 무작위로 30개를 뽑습니다.
    sample = np.random.uniform(0, 1, n_size)
    # 30개의 평균을 계산하여 리스트에 저장합니다.
    sample_means.append(np.mean(sample))

# 3. 시각화
plt.figure(figsize=(10, 6))

# 히스토그램과 밀도 곡선(KDE) 그리기
sns.histplot(sample_means, kde=True, bins=15, color='orange', edgecolor='black')

# 원래 균등 분포의 평균인 0.5 지점에 수직선 표시
plt.axvline(x=0.5, color='red', linestyle='--', label='Theoretical Mean (0.5)')

plt.title(f'CLT: Distribution of {n_iterations} Sample Means (n={n_size})')
plt.xlabel('Sample Mean Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', alpha=0.3)

plt.show()

# 통계치 출력
print(f"100개 평균값들의 평균: {np.mean(sample_means):.4f}")
print(f"100개 평균값들의 표준편차: {np.std(sample_means):.4f}")
```

```python
import gradio as gr
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

def bayesian_update(true_p, total_trials):
    # 1. 실제 데이터 생성 (이항 분포)
    heads = np.random.binomial(total_trials, true_p)
    tails = total_trials - heads

    # 2. 분포 계산을 위한 x축 (확률 0~1)
    x = np.linspace(0, 1, 500)

    # 3. 베이지안 업데이트 (Prior: Beta(1,1) 균등분포 가정)
    # 사후 분포: Beta(α + heads, β + tails)
    y_prior = beta.pdf(x, 1, 1)
    y_posterior = beta.pdf(x, 1 + heads, 1 + tails)

    # 4. 시각화
    plt.figure(figsize=(10, 6), facecolor='#121212')
    plt.plot(x, y_prior, label='Prior (Initial Belief)', color='gray', linestyle='--')
    plt.plot(x, y_posterior, label=f'Posterior (After {total_trials} tosses)', color='#FF4B00', linewidth=3)
    plt.fill_between(x, y_posterior, color='#FF4B00', alpha=0.2)

    plt.axvline(true_p, color='#00FF00', linestyle=':', label=f'True p: {true_p}')

    plt.title(f"Bayesian Learning: {heads}H, {tails}T", color='white', fontsize=15)
    plt.legend()
    plt.grid(alpha=0.2)
    plt.xlim(0, 1)

    # 스타일 설정
    ax = plt.gca()
    ax.set_facecolor('#121212')
    ax.tick_params(colors='white')
    for spine in ax.spines.values():
        spine.set_color('white')

    return plt.gcf()

# Gradio 인터페이스 구성
with gr.Blocks() as demo:
    gr.Markdown("## 🧐 베이지안 업데이트 시뮬레이터")
    gr.Markdown("데이터가 쌓일수록 우리의 믿음이 어떻게 진실에 가까워지는지 확인해보세요.")

    with gr.Row():
        with gr.Column():
            true_p = gr.Slider(0, 1, value=0.7, step=0.01, label="실제 동전 앞면 확률 (True p)")
            trials = gr.Slider(1, 1000, value=100, step=1, label="동전 던지기 횟수 (Total Trials)")
            submit_btn = gr.Button("Submit", variant="primary")

        with gr.Column():
            plot_output = gr.Plot()

    submit_btn.click(bayesian_update, inputs=[true_p, trials], outputs=plot_output)

demo.launch()
```
